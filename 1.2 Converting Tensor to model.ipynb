```python
import torch
from torch import nn
```

```python
x = torch.tensor([
    [100.0, 2.0],
    [200.0, 3.0]]
)
```

### we need to create the model like following

y_pred = w1 * x1 + w2 * x2 + ...+ wn* xn + b 

```python
model = nn.Linear(2, 1)
```

```python
print(model)
```

```python
model.bias = nn.Parameter(torch.tensor([16.0]))
model.weight = nn.Parameter(torch.tensor([[0.2, 0.2]]))
```

```python
y_pred = model(x)
print(y_pred)
```
