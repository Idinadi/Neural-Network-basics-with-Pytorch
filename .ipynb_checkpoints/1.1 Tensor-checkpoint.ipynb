```python
import torch
```

```python
# y_pred = w * x + 1 * b -> Single Neuron

b = torch.tensor(16) # bias term
w = torch.tensor(0.2) # Initialized weights
x = torch.tensor(100) # Input feature
```

```python
y_pred = w * x + b
print(y_pred)
```

### the above equation can be rewrittern until and unless the input feature is a tensor

```python
# You can rewrite the same neuron as

b, w = 16, 0.2
x = torch.tensor(100)
y_pred = x * w + 1 * b
print(y_pred)
```

### For multiple examples which has a single feature then the neuron will be like

```python
b, w = 16, 0.2
x = torch.tensor([100, 200, 300])
y_pred = x * w + 1 * b
print(y_pred)
```

### to view the size of the tensor you can use either shape or size() function

```python
y_pred.shape
```

```python
y_pred.size()
```

# To unpack it, we just need to use the item function to convert a single value

```python
print("This is the original tensor value : ",y_pred[0])
print("This is the unpacked value : ", y_pred[0].item())
```

#### if we have multiple features with multiple examples

```python
b, w = 16, 0.2
x = torch.tensor([
    [100, 1],
    [200, 2],
    [300, 3]]
    )
y_pred = w * x + 1 * b
print(y_pred)
```

### to get value from second row, from all the columns

```python
y_pred[1, :]
```

### Similarly to get value from all the rows and from 1st column

```python
y_pred[:, 1]
```
