```python
import torch
```

```python
x = torch.tensor(3)
```

```python
y  = x * 2
print(y)
```

```python
y.backward()
```

```python
x = torch.tensor(3.0, requires_grad=True)
```

```python
y  = x * 2
print(y)
```

```python
y.backward() #dy/dx = 2
print(x.grad)
```

```python
y.grad #y is not a leaf tensor hence grad is not calculated
```

```python
x = torch.tensor(3.0, requires_grad=True)
```

```python
y  = x ** 2
print(y)
```

```python
y.backward() #dy/dx = 2*3 = 6
print(x.grad)
```
